10. Log File Analyzer
Develop a Log File Analyzer for simulated system logs.
The program should parse text-based log entries, categorize events, count error types,
analyze user activity, and filter logs based on time or severity.
String splitting, pattern matching, and dictionary-based aggregation must be used. The
application should present analytical summaries via a menu-driven interface.

```python

raw_logs = [
    "[INFO] [user_jose] [SYSTEM]: System boot completed",
    "[ERROR] [user_christo] [AUTH]: Database timeout",
    "[ERROR] [user_jose] [AUTH]: Password failed",
    "[WARNING] [user_jose] [DISK]: Disk usage at 85%",
    "[CRITICAL] [admin] [NETWORK]: DDoS pattern detected",
    "[ERROR] [user_christo] [DEPS]: Out of memory error"
]


def parse_logs():
    parsed_list = []
    for line in raw_logs:

        parts = line.split("] ")
        parsed_list.append({
            "severity": parts[0].replace("[", ""),
            "user": parts[1].replace("[", ""),
            "message": parts[2].split(": ")[1],
            "raw": line
        })
    return parsed_list


def show_metrics(logs):
    severity_counts = {}
    user_activity = {}

    for log in logs:

        sev = log["severity"]
        severity_counts[sev] = severity_counts.get(sev, 0) + 1


        user = log["user"]
        user_activity[user] = user_activity.get(user, 0) + 1

    print("\n--- SEVERITY COUNTS ---")
    for k, v in severity_counts.items():
        print(f"{k}: {v}")

    print("\n--- USER ACTIVITY ---")
    for k, v in user_activity.items():
        print(f"{k}: {v} logs")


def filter_logs(logs, keyword):
    print(f"\n--- FILTER RESULTS FOR '{keyword}' ---")
    found = False
    for log in logs:
        if keyword.lower() in log["message"].lower() or keyword.lower() in log["severity"].lower():
            print(log["raw"])
            found = True
    if not found:
        print("No matching log entries found.")


parsed_data = parse_logs()
while True:
    print("\n1. Metrics Summary  2. Search/Filter  3. Exit")
    choice = input("Select option: ").strip()

    if choice == "1":
        show_metrics(parsed_data)
    elif choice == "2":
        filter_logs(parsed_data, input("Enter search keyword: "))
    elif choice == "3":
        print("Exiting Analyzer.")
        break
    else:
        print("Invalid choice.")
```
